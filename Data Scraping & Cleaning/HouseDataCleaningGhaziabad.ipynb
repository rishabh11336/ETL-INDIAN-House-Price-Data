## Filling nan values for rekki

```python
carpetArea = SimpleImputer(strategy='constant', fill_value=0)
carpetArea = carpetArea.fit_transform(pd.DataFrame(house_data['carpetArea']))
house_data['carpetArea'] = pd.DataFrame(carpetArea, columns=['carpetArea'])
```

```python
coveredArea = SimpleImputer(strategy='constant', fill_value=0)
coveredArea = coveredArea.fit_transform(pd.DataFrame(house_data['coveredArea']))
house_data['coveredArea'] = pd.DataFrame(coveredArea, columns=['coveredArea'])
```

```python
sqFtPrice = SimpleImputer(strategy='constant', fill_value=0)
sqFtPrice = sqFtPrice.fit_transform(pd.DataFrame(house_data['sqFtPrice']))
house_data['sqFtPrice'] = pd.DataFrame(sqFtPrice, columns=['sqFtPrice'])
```

## Average percentage for carpet and covered area

```python
len(house_data)
```

```python
house_data['areaPercent'] = 0
```

```python
for i in range(len(house_data)):
    if house_data['coveredArea'][i]!=0 and house_data['carpetArea'][i]!=0:
        house_data['areaPercent'][i] = (int(house_data['coveredArea'][i])-int(house_data['carpetArea'][i]))/int(house_data['coveredArea'][i])
```

```python
house_data[house_data['areaPercent']!=0]['areaPercent'].mean()
```

## fill na carpet area with 80% of covered area

```python
for i in range(len(house_data)):
    if house_data['coveredArea'][i]!=0 and house_data['carpetArea'][i]==0:
        house_data['carpetArea'][i] = int(house_data['coveredArea'][i])*0.8
```
